# Import Library

```python
# 1. Import pustaka
import pandas as pd
from google.colab import files
```

# Baca Data

```python
# 3. Load semua dataset
df_fish = pd.read_csv("/content/fish_dataset.csv")
df_categories = pd.read_csv("/content/fish_categories_dataset.csv")
df_locations = pd.read_csv("/content/fish_locations_dataset.csv")
df_fishers = pd.read_csv("/content/fishers_dataset.csv")
df_prices = pd.read_csv("/content/fish_market_prices_dataset.csv")
```

# Data Cleaning

```python
# 1. Cek missing values untuk fish_dataset
missing_values = df_fish.isnull().sum().to_frame().T
missing_values
```

```python
median_length = df_fish['length_cm'].median()
df_fish['length_cm'] = df_fish['length_cm'].fillna(median_length)
df_fish.head()
```

```python
modus_fish_name = df_fish['fish_name'].mode()[0]
df_fish['fish_name'] = df_fish['fish_name'].fillna(modus_fish_name)
df_fish.head()
```

```python
# 2.1 Isi missing value di kolom weight_kg (numeric) dengan median
median_weight = df_fish['weight_kg'].median()
df_fish['weight_kg'] = df_fish['weight_kg'].fillna(median_weight)
df_fish.head()
```

```python
# 2.3 Isi missing value di kolom catch_date (tanggal) dengan modus
modus_date = df_fish['catch_date'].mode()[0]
df_fish['catch_date'] = df_fish['catch_date'].fillna(modus_date)
df_fish.head()
```

```python
# 2.4 Isi missing value di kolom freshness_score (bisa numeric atau kategori)
# Kalau numerik  median; kalau kategori  modus. Misal kita pakai modus dulu:
modus_freshness = df_fish['freshness_score'].mode()[0]
df_fish['freshness_score'] = df_fish['freshness_score'].fillna(modus_freshness)
df_fish.head()
```

```python
# Cek kembali apakah masih ada missing values
df_fish.isnull().sum()
```

```python
# 3. Hapus data duplikat berdasarkan fish_name
fish_df= df_fish.drop_duplicates(subset='fish_name')
fish_df.head()
```

# Cleaning Categori

```python
# 1. Cek missing values untuk categories
missing_values = df_categories.isnull().sum().to_frame().T
missing_values
```

# Cleaning Market Price

```python
# Cek missing values
missing_market = df_prices.isnull().sum().to_frame().T
missing_market
```

# Cleaning Location

```python
# Cek missing values
missing_market = df_locations.isnull().sum().to_frame().T
missing_market
```

# Cleaning Fisher

```python
# Cek missing values
missing_market = df_fishers.isnull().sum().to_frame().T
missing_market
```

# Join Operation

```python
# Langkah 1: Lakukan inner join antara fish dan fish_categories berdasarkan category_id
joined_df = pd.merge(df_fish, df_categories, on='category_id', how='inner')
joined_df.head()
```

```python
# Langkah 2: Inner join dengan fish_locations berdasarkan location_id
joined_df = pd.merge(joined_df, df_locations, on='location_id', how='inner')
joined_df.head()
```

```python
# Langkah 3: Inner join dengan fishers berdasarkan fisher_id
joined_df = pd.merge(joined_df, df_fishers, on='fisher_id', how='inner')
joined_df.head()
```

# DATA FRAME

# 1. Ratac - rata berat/panjang ikan per kategori

```python
# Hitung rata rata berat/panjang ikan per kategori
rata_rata = fish_df.groupby(['fish_name', 'category_id'])[['weight_kg', 'length_cm']].mean().reset_index()
rata_rata
```

# 2. Total tangkapan Nelayan
