# 1.3 Sampel Data Setiap Kolom

```python
import pandas as pd

# Contoh DataFrame
df = pd.read_csv('C:/TUGAS SEMESTER 3/Pemrograman Lanjut/cards_data.csv')

df = pd.DataFrame(df)

# Fungsi untuk mengambil sample data dari setiap kolom
def sample_per_column(df, n=1):
    sample_data = {}
    for column in df.columns:
        sample_data[column] = df[column].sample(n=n, random_state=42).tolist()
    return sample_data

# Mendapatkan sample data (1 nilai dari setiap kolom)
sample_data = sample_per_column(df, n=1)

# Menampilkan hasil
print("Sample data dari setiap kolom:")
for column, sample in sample_data.items():
    print(f"{column}: {sample}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca data CSV
df = pd.read_csv('C:/TUGAS SEMESTER 3/Pemrograman Lanjut/cards_data.csv')

# Fungsi untuk analisis korelasi
def analisis_korelasi(dataframe):
    # 1. Mengambil hanya kolom numerik untuk analisis korelasi
    dataframe_numerik = dataframe.select_dtypes(include=['number'])
    
    # 2. Menghitung korelasi
    korelasi = dataframe_numerik.corr()
    
    # 3. Membuat visualisasi heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(korelasi, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
    plt.title('Matriks Korelasi')
    plt.show()
    
    # 4. Membuat scatter plot untuk semua pasangan variabel
    plt.figure(figsize=(12, 10))
    sns.pairplot(dataframe_numerik)
    plt.suptitle('Scatter Plot Matrix', y=1.02)
    plt.show()
    
    # 5. Mencetak interpretasi korelasi
    print("\nInterpretasi Korelasi:")
    for col1 in dataframe_numerik.columns:
        for col2 in dataframe_numerik.columns:
            if col1 < col2:  # Menghindari pengulangan
                corr = korelasi.loc[col1, col2]
                interpretasi = interpret_correlation(corr)
                print(f"{col1} dengan {col2}: {corr:.3f} ({interpretasi})")

# Fungsi untuk interpretasi nilai korelasi
def interpret_correlation(value):
    abs_value = abs(value)
    if abs_value >= 0.8:
        return "Korelasi sangat kuat"
    elif abs_value >= 0.6:
        return "Korelasi kuat"
    elif abs_value >= 0.4:
        return "Korelasi sedang"
    elif abs_value >= 0.2:
        return "Korelasi lemah"
    else:
        return "Korelasi sangat lemah"

# Menjalankan analisis
print("Analisis Korelasi:")
print("==================")
analisis_korelasi(df)
```
